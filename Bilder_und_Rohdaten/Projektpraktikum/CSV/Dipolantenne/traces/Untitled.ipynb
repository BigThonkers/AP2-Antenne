```python
import csv

anttrimm2 = []
anttrimm  = []
ant22     = []
ant12     = []

liste = [anttrimm2, anttrimm, ant22, ant12]
names = ['ant trimm2.csv', 'ant trimm.csv', 'ant2.2.csv', 'ant1.2.csv']

for l in range(len(liste)):

    with open(names[l], 'r') as csvFile:
        reader = csv.reader(csvFile)
        i = 0
        for row in reader:
            i += 1
            if i > 2:
                liste[l].append(np.array([float(row[0]), float(row[2])]))
                #print(row)
    #           
    csvFile.close()

#

print(np.array(anttrimm2))
#print(anttrimm)
```

```python
plt.rcParams.update({'font.size': 17})
plt.figure(figsize=(15,9))
plt.title("Vergleich der Antennen vor und nach Trimmen")


xantt2 = np.array(anttrimm2)[:,0]
yantt2 = np.array(anttrimm2)[:,1]

xantt  = np.array(anttrimm )[:,0]
yantt  = np.array(anttrimm )[:,1]

xant22 = np.array(ant22    )[:,0]
yant22 = np.array(ant22    )[:,1]

xant12 = np.array(ant12    )[:,0]
yant12 = np.array(ant12    )[:,1]

#plt.plot(xantt , yantt )

plt.plot(xant12, yant12, color="#ff7f0e", label='1. Antenne')
plt.plot(xant22, yant22, color="#1f77b4", label='2. Antenne vor trimmen')
plt.plot(xantt2, yantt2, color="#2ca02c", label='2. Antenne getrimmt')

plt.plot([1.616e9, 1.616e9], [-35,5],"o-",color="red", label="Mess-Frequenz: 1,616 GHz")

plt.plot(xant12[303], yant12[303], "o",  color="#ff7f04",markeredgecolor="black")
plt.plot(xant22[303], yant22[303], "o",  color="#1f77b4",markeredgecolor="black")
plt.plot(xantt2[303], yantt2[303], "o",  color="#2ca02c",markeredgecolor="black")

plt.ylim([-30,0])
plt.xlim([1e9,2e9])

#plt.semilogy()

plt.legend(loc=4)

plt.minorticks_on()
plt.grid(which='minor', linestyle=':', color='dimgray')
plt.grid(color='gray')

plt.xlabel('Frequenz in [Hz]')
plt.ylabel('Intensitt in [dBm]')

plt.savefig(filename='Antennentrimmen.jpg')

plt.show()
```
